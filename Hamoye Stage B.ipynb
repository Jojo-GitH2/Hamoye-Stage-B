```python
data = data.drop(["date","lights"], axis = 1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
normal_data = pd.DataFrame(scaler.fit_transform(data), columns = data.columns)
```

```python
data_x = normal_data.drop( columns = "Appliances")
```

```python
data_y = normal_data["Appliances"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data_x, data_y, random_state = 42, test_size = 0.3)
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
regr_mod = LinearRegression()
regr_mod.fit(x_train, y_train)
pred_data_y = regr_mod.predict(x_test)

pred_data_y
```

```python
MAE = mean_absolute_error(y_test, pred_data_y)
```

```python
round(MAE,2)
```

```python
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, pred_data_y))
```

```python
rmse
```

```python
RRS = np.square(rmse) * len(y_test)
```

```python
round(RRS, 2)
```

```python
round(rmse, 3)
```

```python
from sklearn.metrics import r2_score

coef_det = r2_score(y_test, pred_data_y)
```

```python
round(coef_det, 2)
```
